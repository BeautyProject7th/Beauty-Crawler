```python
import requests
import time
import pandas as pd
from pandas import DataFrame
from bs4 import BeautifulSoup
```

```python
url = 'http://www.thefaceshop.com'
main_url = url + '/mall/product/category.jsp?cate_seq=7'
source_code = requests.get(main_url, allow_redirects=False)
plain_text = source_code.text
soup = BeautifulSoup(plain_text,'html.parser')
```

```python

div_list = []
cate_list = []
cate_link_list = []
sub_cate_list = []
cate_seq_list = []
pdName_list = []
pdImg_list = []
price_list = []
pdLink_list = []
pdType_list = []

dic = {}
```

```python
def find_div():
    i = 0
    for p_div in soup.findAll('li',{'class':'ja_gnbover'}) :
        if i < 3 :
            div = p_div.find('img').get('alt')
            dic["div"] = div
            print(div)
            i += 1
            find_cate(p_div)
```

```python
def find_cate_seq(link):
    index = link.find("=")
    return link[index+1:]
```

```python
def find_cate(p_div):
    temp = p_div.find('ul',{'class':'GL1'})
    for p_cate in temp.findAll('li'):
        cate = p_cate.find('a')
        dic["cate"] = cate.text
        dic["cate_link"] = cate.get('href')
        print(cate.text)
        
        if dic["div"] == '남성라인' :
            dic["sub_cate"] = ""
            dic["cate_seq"] = find_cate_seq(cate.get('href'))
            move_pdPage(cate.get('href'))
        else : 
            move_sub_cate(p_div.find('div',{'class':'GL2'}))
```

```python
def move_sub_cate(p_cate):
    for p_sub in p_cate.findAll('a'):
        sub_cate = p_sub.text
        dic["sub_cate"] = sub_cate
        print(sub_cate)
            
        sub_link = p_sub.get('href')
        dic["cate_seq"] = find_cate_seq(sub_link)
        move_pdPage(sub_link)
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def move_pdPage(pd_link):
    time.sleep(1)
    if __name__ == "__main__":
        driver = webdriver.Firefox()
        pd_url = url + pd_link
        driver.get(pd_url)
        time.sleep(1)
        soup2 = BeautifulSoup(driver.page_source,'html.parser')
        
        pdCount = soup2.find('div',{'class':'result'}).select('span')
        p_soup = soup2
        while int(pdCount[0].text) > int(pdCount[1].text):
            driver.execute_script("getProdAddList('add');return false;")
            time.sleep(3)
            p_soup = BeautifulSoup(driver.page_source,'html.parser')
            pdCount = p_soup.find('div',{'class':'result'}).select('span')
        
        driver.quit()
        get_product(p_soup.find('ul',{'class':'prdt_list'}))
```

```python
def get_product(soup2):
    for product in soup2.findAll('li'):
        pdImg = product.find('div',{'class':'img'}).find('img').get('src')
        pdName = product.find('p',{'class':'prdt_name_1'}).text.strip()
        pdLink = product.find('a',{'class':'detail'}).get('href')
        price = product.find('span',{'class':'prdt_price_1'}).text.strip()
        pdType = ""
        for t in product.find('p',{'class':'prdt_type_1'}).select('span'):
            pdType += t.text.strip()
        
        div_list.append(dic["div"])
        cate_list.append(dic["cate"])
        cate_link_list.append(dic["cate_link"])
        sub_cate_list.append(dic["sub_cate"])
        cate_seq_list.append(dic["cate_seq"])
        pdName_list.append(pdName)
        pdImg_list.append(pdImg)
        pdLink_list.append(pdLink)
        price_list.append(price)
        pdType_list.append(pdType)
        print(pdName)
```

```python
result = {
    'div' : div_list,
    'cate' : cate_list,
    'cate_link' : cate_link_list,
    'sub_cate' : sub_cate_list,
    'cate_seq' : cate_seq_list,
    'pdName' : pdName_list,
    'pdImg' : pdImg_list,
    'price' : price_list,
    'pdLink' : pdLink_list,
    'pdType' : pdType_list
}
```
