```python
import requests
import time
import pandas as pd
from pandas import DataFrame
from bs4 import BeautifulSoup
```

```python
url = 'http://www.laneige.com'
main_url = url + '/kr/ko/product/new_best/new.html'
source_code = requests.get(main_url, allow_redirects=False)
plain_text = source_code.text
soup = BeautifulSoup(plain_text,'html.parser')
```

```python
cate_list = []
div_list = []
sub_cate_list = []
sub_link_list = []
pdName_list = []
pdImg_list = []
price_list = []
pdLink_list = []

dic = {}
```

```python
def find_cate():
    temp = soup.find('div',{'class':'dep2wrap Product'})
    i = 1
    for p_cate in temp.select('li'):
        print("i : "+str(i))
        if i>1 and i<5:
            cate = p_cate.select('a')[0].text.strip()
            dic["cate"] = cate
            print(cate)
            find_sub_cate(p_cate)
        i += 1
```

```python
def find_sub_cate(p_cate):
    temp = p_cate.find('div',{'class':'dep3wrap'})
    
    for p_div in temp.findAll('dl'):
        div = p_div.select('dt > a')[0].text
        print("div:"+div)
        dic["div"] = div
        for p_sub in p_div.findAll('dd'):
            finda = p_sub.find('a')
            sub = finda.text
            sub_link = finda.get('href')
            print(sub)
            print(sub_link)
            dic["sub_cate"] = sub
            dic["sub_link"] = sub_link
            move_pdPage(sub_link)
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def move_pdPage(sub_link):
    time.sleep(1)
    if __name__ == "__main__":
        driver = webdriver.Firefox()
        driver.get(sub_link)
        soup2 = BeautifulSoup(driver.page_source,'html.parser')
        time.sleep(3)
        
        flag = 1
        while flag==1:
            btn = driver.find_element_by_class_name('circlebtn.more')
            if btn.is_displayed():
                print("Element found")
                btn.click()
                time.sleep(2)
            else:
                print("Element not found")
                flag = 0
            
        p_soup = BeautifulSoup(driver.page_source,'html.parser')
        pd_list = p_soup.find('div',{'class':'product-list-wrap opt-borderbot'})
        driver.quit()
        get_product(pd_list)
```

```python
def get_product(soup2):
    for p in soup2.findAll('div',{'class':'product-list-item'}):
        product = p.findAll('a')[0]
        
        pdImg = url + product.find('img').get('src')
        pdName = product.find('span',{'class':'product-list-title opt-btb opt-tc opt-text-arbfont'}).text
        pdLink = product.get('href')
        price = product.find('span',{'class':'product-list-price opt-btb opt-tc opt-text-arbfont'}).text[:-1]
        
        div_list.append(dic["div"])
        cate_list.append(dic["cate"])
        sub_cate_list.append(dic["sub_cate"])
        sub_link_list.append(dic["sub_link"])
        pdName_list.append(pdName)
        pdImg_list.append(pdImg)
        pdLink_list.append(pdLink)
        price_list.append(price)
        print(pdName)
```

```python
result = {
    'div' : div_list,
    'cate' : cate_list,
    'sub_cate' : sub_cate_list,
    'sub_cate_link' : sub_link_list,
    'pdName' : pdName_list,
    'pdImg' : pdImg_list,
    'price' : price_list,
    'pdLink' : pdLink_list
}
```
