## 1-4. subCategory 확인하기

```python
findCate = soup.find_all("ul", {"id" : flag})
```

```python
findCate = str(findCate)
findCate = findCate.replace("><", ">\n<")
findCate = findCate.split("\n")
```

```python
findCate
```

```python
subCate = []
for line in findCate:
    if "a href" in line:
        subCate.append(str(line))
```

```python
subCate
```

```python
totalSub = []
for line in subCate:
    cate_id = line[36:46]
    cate_name = line[50:-4]
    id_name = [cate_id, cate_name]
    totalSub.append(id_name)
```
