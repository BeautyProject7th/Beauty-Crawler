```python
import requests
from bs4 import BeautifulSoup
```

## wiki를 활용하여 화장품 회사명 가져오기

```python
wikiurl = 'https://ko.wikipedia.org/wiki/한국화장품'
wiki_page = requests.get(wikiurl)
wiki_page_code = wiki_page.text
wiki = BeautifulSoup(wiki_page_code, 'lxml')
```

### 전체 페이지

```python
wiki
```

### 특정 부분

```python
cosmetic_wiki = wiki.findAll('table', {"class":"navbox collapsible autocollapse nowraplinks"})
```

```python
cosmetic_wiki
```

### 특정 부분의 텍스트만

```python
wiki_text = cosmetic_wiki[0].text
print(wiki_text)
```

#### 텍스트에서 쓸모없는 앞부분 삭제

```python
wiki_text = wiki_text[27:]
print(wiki_text)
```

#### 텍스트에서 한국 화장품만(외국 화장품 제외)

```python
korea_cosmetic = wiki_text[ :842]
print(korea_cosmetic)
```

#### 쓸데없는 공백문자 정리

```python
korea_cosmetic = korea_cosmetic.replace("\n", "/")
korea_cosmetic
```

```python
korea_cosmetic = korea_cosmetic.replace("//////", "\n")
print(korea_cosmetic)
```

```python
korea_cosmetic = korea_cosmetic.replace("KT&G/////코스모코스", "KT&G-코스모코스")
korea_cosmetic = korea_cosmetic.replace("KGC라이프앤진", "KT&G-라이프앤진")
korea_cosmetic = korea_cosmetic.replace("////에이블씨앤씨", "에이블씨앤씨")
print(korea_cosmetic)
```

```python
korea_cosmetic = korea_cosmetic.replace("////", ":")
print(korea_cosmetic)
```

```python
korea_cosmetic = korea_cosmetic.replace("/", ",")
print(korea_cosmetic)
```

```python
korea_cosmetic = korea_cosmetic.replace("\n", "\n\n")
print(korea_cosmetic)
```

```python
korea_cosmetic = korea_cosmetic.replace("\n\n", "\n\n\n")
korea_cosmetic = korea_cosmetic.replace(":", "\n\n")
korea_cosmetic = korea_cosmetic.replace(",", "\n")
```
