```python
item = playlist_bs.find_all("a", {"class" : "yt-uix-sessionlink yt-uix-tile-link spf-link yt-ui-ellipsis yt-ui-ellipsis-2"})
```

```python
from time import sleep
```

```python
itemdf = []
itemCol1 = ["유튜버", "분류명", "분류주소", "컨텐츠명", "컨텐츠주소", "썸네일"]
item_address = ""
item_ilist = ""
#content_cnt = ""
content_name = ""
content_site = ""
content_thumbnail = ""
content_movie = ""
content_more = ""

for line in item:
    item_address = str(line)
    item_address = item_address.split(" ")
    item_address = item_address[9]
    item_address = item_address[6:-1]
    item_address = youtube+item_address
    
    item_ilist = str(line)
    item_ilist = item_ilist.split("\">")
    item_ilist = item_ilist[1]
    item_ilist = item_ilist[:-4]
    if "amp;" in item_ilist:
        item_ilist = item_ilist.replace("amp;", "")
        
# #     driver.get(item_address)
# #     content_cnt = driver.page_source
# #     content_cnt = BeautifulSoup(content_cnt)
# #     content_cnt = content_cnt.find_all("ul", {"class" : "pl-header-details"})
# #     content_cnt = str(content_cnt)
# #     content_cnt = content_cnt.split("<li>")
# #     content_cnt = content_cnt[2]
# #     content_cnt = content_cnt[4:-6]
# #     content_cnt = int(content_cnt)
    
    driver.get(item_address)
    sleep(3)
    content = driver.page_source
    content = BeautifulSoup(content)
        
    content_site = content.find_all("a", {"class" : "pl-video-title-link yt-uix-tile-link yt-uix-sessionlink spf-link "})
    content_site = list(content_site)
    content_site_cnt = len(content_site)
    content_thumbnail = content.find_all("span", {"class" : "video-thumb yt-thumb yt-thumb-72"})
    content_thumbnail = list(content_thumbnail)
    
    for i in range(content_site_cnt):
        test_site = str(content_site[i])
        test_site = test_site.split("\n")
        content_name = test_site[1]
        test_site = test_site[0]
        content_name = content_name.replace("      ", "")
        test_site = test_site.split(" ")
        test_site = test_site[8]
        test_site = test_site[6:-2]
        test_site = youtube+test_site

        test_thumbnail = content_thumbnail[i]
        test_thumbnail = str(test_thumbnail)
        test_thumbnail = test_thumbnail.split("\n")
        test_thumbnail = test_thumbnail[3]
        test_thumbnail = test_thumbnail.split(" ")
        same = 0
        for line in test_thumbnail:
            if "data-thumb" in line:
                same = +1
        if same == 0:
            test_thumbnail = test_thumbnail[5]
            test_thumbnail = test_thumbnail[5:-1]
        if same == 1:
            test_thumbnail = test_thumbnail[3]
            test_thumbnail = test_thumbnail[12:-1]

        itemAll = [youtuber, item_ilist, item_address, content_name, test_site, test_thumbnail]
        itemdf.append(itemAll)
        print(itemAll)
```
