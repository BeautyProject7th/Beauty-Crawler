```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Firefox()
```

```python
import time
```

```python
total = []
for i in range(9):
    channel_search = "https://www.youtube.com/channels?q=%EB%B7%B0%ED%8B%B0&page="
    channel_search = channel_search + str(i+1)
    channel_page = driver.get(channel_search)
    channel_html = driver.page_source
    time.sleep(4)
    channel_bs = BeautifulSoup(channel_html)
    menu = channel_bs.find_all("div", {"class" : "yt-gb-shelf-hero yt-gb-compact-shelf"})
    temp = str(menu)
    menu = temp
    menu = menu.split("\n")
    cnt = len(menu)
    address = ""
    for i in range(cnt):
        test = menu[i]
        if "yt-gb-shelf-hero-thumb" in test:
            address = test
            total.append(str(address))
```

```python
total
```

```python
total_cnt = len(total)
```

```python
Tsum = []
for i in range(total_cnt):
    result = ""
    juso = ""
    result = total[i]
    result = result[40:]
    result = result.split("\" ")
    juso = result[0]
    name = result[1]
    name = name[7:-2]
    sumsum = [juso, name]
    print(sumsum)
    Tsum.append(sumsum)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(Tsum, columns = ["juso", "name"])
```
