```python
import requests
import time
import pandas as pd
from pandas import DataFrame
from bs4 import BeautifulSoup
```

```python
url = 'http://www.hera.com'
main_url = url + '/kr/ko/product/make-up.all.all.html'
source_code = requests.get(main_url, allow_redirects=False)
plain_text = source_code.text
soup = BeautifulSoup(plain_text,'html.parser')
```

```python
div_list = []
div_link_list = []
cate_list = []
sub_cate_list = []
sub_link_list = []
pdName_ko_list = []
pdName_en_list = []
pdImg_list = []
pdLink_list = []

dic = {}
```

```python
def find_div():
    temp = soup.find('ul',{'class':'product-select-list'})
    for p_div in temp.findAll('a'):
        div = p_div.find('span').text
        div_link = p_div.get('href')
        
        dic["div"] = div
        dic["div_link"] = div_link
        
        print(div)
        time.sleep(1)
        div_url = url + div_link
        sc = requests.get(div_url, allow_redirects=False)
        soup2 = BeautifulSoup(sc.text,'html.parser')
        find_cate(soup2)
```

```python
def find_cate(soup2):
    i = 1
    temp = soup2.find('div',{'class':'product-select-box'}) 
    for p_cate in soup2.findAll('div',{'class':'product-select'}):
        if i > 1:
            cate = p_cate.find('h3').text
            dic["cate"] = cate
            print(cate)
            if cate == "ETC" : move_sub_cate(p_cate,0)
            else : move_sub_cate(p_cate,1)
        i += 1
```

```python
def move_sub_cate(p_cate,all_flag):
    cate_index = 1
    for p_sub in p_cate.findAll('a'):
        if cate_index > all_flag : #전체 제외
            #sub카테고리 이름 다듬기
            sub_cate = p_sub.find('span').text
            dic["sub_cate"] = sub_cate
            print(sub_cate)
            
            sub_link = p_sub.get('href')
            dic["sub_link"] = sub_link
            print(sub_link)
            
            move_sub_page(sub_link)
        cate_index += 1
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def move_sub_page(sub_link):
    time.sleep(1)
    if __name__ == "__main__":
        driver = webdriver.Firefox()
        sub_url = url + sub_link
        driver.get(sub_url)
        soup2 = BeautifulSoup(driver.page_source,'html.parser')

        #자동 스크롤 임시로 10번 하게 함
        for i in range(0,10):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
        
        p_soup = BeautifulSoup(driver.page_source,'html.parser')
        driver.close()
        get_product(p_soup)
```

```python
def get_product(soup2):
    for product in soup2.findAll('a',{'class':'product-item'}):
        pdImg = product.find('img').get('src').strip()
        pdName_ko = product.find('b').text
        pdName_en = product.find('i').text
        pdLink = product.get('href')
        
        div_list.append(dic["div"])
        div_link_list.append(dic["div_link"])
        cate_list.append(dic["cate"])
        sub_cate_list.append(dic["sub_cate"])
        sub_link_list.append(dic["sub_link"])
        pdName_ko_list.append(pdName_ko)
        pdName_en_list.append(pdName_en)
        pdImg_list.append(pdImg)
        pdLink_list.append(pdLink)
        
        print(pdName_ko)
```

```python
result = {
    'div' : div_list,
    'div_link' : div_link_list,
    'cate' : cate_list,
    'sub_cate' : sub_cate_list,
    'sub_cate_link' : sub_link_list,
    'pdName_ko' : pdName_ko_list,
    'pdName_en' : pdName_en_list,
    'pdImg' : pdImg_list,
    'pdLink' : pdLink_list
}
```

```python
df = pd.DataFrame(result)
df.to_csv('/Users/mijeong/Documents/WebCrawler/results/hera.csv', sep=',', na_rep='NaN')
```
