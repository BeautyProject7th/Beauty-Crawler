```python
import requests
import time
import pandas as pd
from pandas import DataFrame
from bs4 import BeautifulSoup
```

```python
url = 'http://www.hera.com/kr/ko/product'
main_url = url + '/make-up.all.all.html'
source_code = requests.get(main_url, allow_redirects=False)
plain_text = source_code.text
soup = BeautifulSoup(plain_text,'html.parser')
```

```python
div_list = []
cate_list = []
cate_link_list = []
catCd01_list = []
sub_cate_list = []
catCd02_list = []
pdName_list = []
pdImg_list = []
price_list = []

dic = {}
```

```python
def find_div():
    soup = soup.find('ul',{'class':'product-select-list'})
    print("a")
    for p_div in soup.findAll('a'):
        div = p_div.find('span').text
        print(div)
        #dic["div"] = div[0].text
        #find_cate(p_div)
```

```python
def find_cate(p_div):
    for p_cate in p_div.select('li > a'):
        print(p_cate.text)
        cate = p_cate.text
        cate_link = p_cate.get('href')
        
        #catCd01따로 가져오기
        index = cate_link.find("catCd01=")
        if index > 0 : 
            catCd01 = cate_link[index+8:]
            catCd01 = catCd01[0]+catCd01[1]
        else : catCd01 = ""
        print(catCd01)
        
        dic["cate"] = cate
        dic["cate_link"] = cate_link
        dic["catCd01"] = catCd01
        
        if dic["div"] == "유형별 제품":
            if dic["cate"] == "세트 상품N":
                dic["cate"] = "세트 상품"
                move_cate(cate_link,0)
            else : move_cate(cate_link,1)
        else : move_cate(cate_link,0)
```

```python
def move_cate(cate_link,all_flag):
    time.sleep(1)
    cate_url = url + cate_link
    print(cate_url)
    sc = requests.get(cate_url, allow_redirects=False)
    soup2 = BeautifulSoup(sc.text,'html.parser')
    
    cate_index = 1
    soup2 = soup2.find('ul',{'class':'lnbList'})
    for p_sub in soup2.findAll('a',{'class':'color'}):
        if cate_index > all_flag : #전체 제외
            #sub카테고리 이름 다듬기
            sub_cate = p_sub.text.strip()
            i = sub_cate.find("(")
            sub_cate = sub_cate[:-(len(sub_cate)-i)]
            print(sub_cate)
            
            sub_link = p_sub.get('href')
            index = sub_link.find("catCd02=")
            catCd02 = sub_link[index+8:]
            print(catCd02)
            
            if dic["div"] == "유형별 제품" : 
                dic["sub_cate"] = sub_cate
            else : dic["sub_cate"] = ""
            dic["catCd02"] = catCd02
            
            move_sub_page(sub_link)
        cate_index += 1
```

```python
def move_sub_page(sub_link):
    page_index=1
    page = 1
    while page_index <= page:
        time.sleep(1)
        sub_url = url+sub_link+"&pageNo="+str(page_index)
        sc = requests.get(sub_url, allow_redirects=False)
        soup2 = BeautifulSoup(sc.text,'html.parser')
        page = len(soup2.findAll('span',{'class':'num'}))+1
        get_product(soup2)
        page_index += 1
```

```python
def get_product(soup2):
    for product in soup2.find('div',{'class':'listStyle1'}).findAll('div',{'class':'listCon'}):
        pdImg = product.find('img').get('src').strip()
        pdName = product.find('img').get('alt').strip()
        price = product.find('p',{'class':'price'}).text.strip()
        price = price[:-(len(price)-price.find("원"))]
        print(pdName + " " + price)
        
        div_list.append(dic["div"])
        cate_list.append(dic["cate"])
        cate_link_list.append(dic["cate_link"])
        catCd01_list.append(dic["catCd01"])
        sub_cate_list.append(dic["sub_cate"])
        catCd02_list.append(dic["catCd02"])
        pdName_list.append(pdName)
        pdImg_list.append(pdImg)
        price_list.append(price)
```

```python
result = {
    'div' : div_list,
    'cate' : cate_list,
    'cate_link' : cate_link_list,
    'catCd01' : catCd01_list,
    'sub_cate' : sub_cate_list,
    'catCd02' : catCd02_list,
    'pdName' : pdName_list,
    'price' : price_list,
    'pdImg' : pdImg_list    
}
```

```python
df = pd.DataFrame(result)
df.to_csv('/Users/mijeong/Documents/WebCrawler/results/innisfree.csv', sep=',', na_rep='NaN')
```
