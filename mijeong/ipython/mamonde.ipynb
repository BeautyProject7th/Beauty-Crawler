```python
import requests
import time
import pandas as pd
from pandas import DataFrame
from bs4 import BeautifulSoup
```

```python
url = 'http://www.mamonde.com'
main_url = url + '/kr/ko/shopping/product/skin_care.html'
source_code = requests.get(main_url, allow_redirects=False)
plain_text = source_code.text
soup = BeautifulSoup(plain_text,'html.parser')
```

```python
div_list = []
cate_list = []
cate_link_list = []
sub_cate_list = []
category2_list = []
pdName_list = []
pdImg_list = []
price_list = []
pdLink_list = []
pdDes_list = []

dic = {}
```

```python
def find_div():
    temp = soup.find('div',{'class':'depth2_group clearfix'})
    for p_div in temp.select('div'):
        div = p_div.find('p').text
        dic["div"] = div
        
        print(div)
        find_cate(p_div)
```

```python
def find_cate(p_div):
    for p_cate in p_div.findAll('li',{'class':'depth2_li'}):
        cate = p_cate.select('a')[0]
        dic["cate"] = cate.text
        dic["cate_link"] = cate.get('href')
        print(cate.text)
        
        if cate.text == '자외선차단' : break
        elif cate.text == '스킨케어' or cate.text == '메이크업' : move_sub_cate(p_cate)
        else : 
            dic["sub_cate"] = ""
            dic["category2"] = ""
            move_pdPage(cate.get('href'))
```

```python
def move_sub_cate(p_cate):
    for p_sub in p_cate.findAll('li',{'class':'depth3_li'}):
        sub_cate = p_sub.find('a').text
        dic["sub_cate"] = sub_cate
        print(sub_cate)
            
        sub_link = p_sub.find('a').get('href')
        i = sub_link.find("=")
        category2 = sub_link[i+1:]
        dic["category2"] = category2
        move_pdPage(sub_link)
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def move_pdPage(pd_link):
    time.sleep(1)
    if __name__ == "__main__":
        driver = webdriver.Firefox()
        pd_url = url + pd_link
        driver.get(pd_url)
        time.sleep(1)
        soup2 = BeautifulSoup(driver.page_source,'html.parser')
        
        for i in range(0,5):
            driver.execute_script("fnShoppingListSearch();trackClicksEx('read more', true);")
            time.sleep(3)
            
        p_soup = BeautifulSoup(driver.page_source,'html.parser')
        driver.quit()
        get_product(p_soup)
```

```python
def get_product(soup2):
    for product in soup2.findAll('div',{'class':'item'}):
        pdImg = product.find('img').get('src').strip()
        pdName = product.find('p',{'class':'opt_tit'}).text
        pdLink = product.find('a').get('href')
        pdDes = product.find('p',{'class':'opt_des'}).text
        price = product.find('span',{'class':'opt_price'}).text[:4]
        
        div_list.append(dic["div"])
        cate_list.append(dic["cate"])
        cate_link_list.append(dic["cate_link"])
        print("aa : " + dic["cate_link"])
        sub_cate_list.append(dic["sub_cate"])
        category2_list.append(dic["category2"])
        pdName_list.append(pdName)
        pdImg_list.append(pdImg)
        pdLink_list.append(pdLink)
        price_list.append(price)
        pdDes_list.append(pdDes)
        print(pdName)
```

```python
result = {
    'div' : div_list,
    'cate' : cate_list,
    'cate_link' : cate_link_list,
    'sub_cate' : sub_cate_list,
    'category2' : category2_list,
    'pdName' : pdName_list,
    'pdImg' : pdImg_list,
    'price' : price_list,
    'pdLink' : pdLink_list,
    'pdDes' : pdDes_list
}
```
