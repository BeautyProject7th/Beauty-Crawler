```python
def move_pdPage(pd_link):
    print(pd_link)
    time.sleep(1)
    if __name__ == "__main__":
        driver = webdriver.Firefox()
        pd_url = url + pd_link
        driver.get(pd_url)
        time.sleep(1)
        soup2 = BeautifulSoup(driver.page_source,'html.parser')
        
        #몇 페이지까지 있는지 확인
        pdCount = soup2.find('div',{'class':'paging'})
        count = pdCount.text.strip()
        count = count[len(count)-1:]
        print(count)
        
        for i in range(1,int(count)+1):
            driver.execute_script("javascript:searchByTarget('"+str(i)+"', '');")
            time.sleep(2)
            p_soup = BeautifulSoup(driver.page_source,'html.parser')
            #pdList = p_soup.find('div',{'class':'NBPrdList1_WrapSmall'})
            for product in p_soup.findAll('li',{'class':'prdList'}):
                get_product(product)
        driver.quit()
```
