# Videos Information [update, description]

````python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from slacker import Slacker
import time
from datetime import datetime 
````

````python
videos_df = pd.read_csv("youtuber_videos_info.csv", encoding = "utf-8")
videos_df = videos_df.fillna(0)
````

````python
youtube = "https://www.youtube.com"
cnt_videos = len(videos_df)
````

````python
cnt_videos
````

````python
def init():
    videos_df["update"] = None
    videos_df["description"] = None
````

````python
def get_address(i):
    address = youtube + videos_df["video_id"][i]
    return address
````

````python
def get_address_bs(ad):
    open_address = driver.get(ad)
    get_source = driver.page_source
    global address_bs
    address_bs = BeautifulSoup(get_source, "html.parser")
    return address_bs
````

````python
def find_upload_date(address_bs):
    upload = address_bs.find_all("div", {"id" : "watch-uploader-info"})
    upload = str(upload)
    upload = upload[63:-16]
    return upload
````

````python
def find_description(address_bs):
    global description
    describe = address_bs.find_all("p", {"id" : "eow-description"})
    describe = str(describe)
    describe = describe[34:-5]
    describe = describe.replace("</a>", "")
    describe = describe.replace("<br/>", "\n")
    describe = describe.split("<a class=\"")
    for i in range(len(describe)):
        if "yt-uix-servicelink" in describe[i]:
            result = describe[i]
            result = result.split("target=\"_blank\">")
            describe[i] = result[1]
        if "yt-uix-sessionlink" in describe[i]:
            result = describe[i]
            result = result.split("\">")
            describe[i] = result[1]
    description = "".join(describe)
    return description
````

````python
Start_time = datetime.now()
driver = webdriver.Firefox()
for i in range(cnt_videos):
    if videos_df["update"][i] == "0":
        try:
            while(1):
                print(i)
                none_num += 1
                slack_check.chat.post_message('#crawling', "Num : " + str(i) + ", https://www.youtube.com" + videos_df["video_id"], as_user=True)
                address = get_address(i)
                address_bs = get_address_bs(address)
                upload = find_upload_date(address_bs)
                description = find_description(address_bs)
                videos_df["update"][i] = upload
                videos_df["description"][i] = description
                slack_check.chat.post_message('#crawling', videos_df["description"][i], as_user=True)
                videos_df.to_csv("youtuber_videos_info.csv", encoding="utf-8", index=False)
                if upload != 0:
                    break
                else:
                    slack.chat.post_message('#finish', str(i)+":0")
        except:
            slack.chat.post_message('#finish', "Error in Video Crawler")
driver.quit()
````
